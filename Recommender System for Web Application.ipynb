```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
book582_user100_pivot = pd.read_csv('dataset/django/python_mba_book582_user100_reviews.csv')
book582_user100_pivot.reset_index(inplace = True)
book582_user100_pivot['Transaction'] = book582_user100_pivot['index']
book582_user100_pivot.set_index('Transaction', inplace=True)
del book582_user100_pivot['index']
```

```python
# create frequent itemsets
itemsets = apriori(book582_user100_pivot, min_support=0.01, use_colnames=True, max_len= 5)
```

<h2>User-to-User</h2>

```python
from collections import defaultdict

from surprise import SVD
from surprise import Dataset, Reader

def get_top_n(predictions, n=10):
    '''Return the top-N recommendation for each user from a set of predictions.

    Args:
        predictions(list of Prediction objects): The list of predictions, as
            returned by the test method of an algorithm.
        n(int): The number of recommendation to output for each user. Default
            is 10.

    Returns:
    A dict where keys are user (raw) ids and values are lists of tuples:
        [(raw item id, rating estimation), ...] of size n.
    '''

    # First map the predictions to each user.
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```

```python
user_book_rating = pd.read_csv('dataset/django/cf_book582_user100_reviews.csv')
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(user_book_rating[['user_id', 'book_id', 'rating']], reader)
trainset = data.build_full_trainset()
algo = SVD()
algo.fit(trainset)
testset = trainset.build_anti_testset()
predictions = algo.test(testset)
top_n = get_top_n(predictions, n=3)

# dump our model
file_name = os.path.expanduser('dataset/django/cf_user_to_user.sav')
dump.dump(file_name, algo=top_n)
# _, loaded_algo = dump.load(file_name)
```
